# Preprocessing

#### EDA

```python
test_df.drop(columns=test_df.columns[0], inplace=True)
```

```python
train_df.drop(columns=train_df.columns[0], inplace=True)
```

```python
test_df.isnull().sum()
```

```python
train_df.isnull().sum()
```

#### Mutlak lahan hutan

```python
train_df['Lahan Hutan']=train_df['Lahan Hutan'].abs()
test_df['Lahan Hutan']=test_df['Lahan Hutan'].abs()
```

#### Memisahkan Negara/Tahun Menjadi 2 Kolom

```python
split_negara=train_df['Negara/Tahun'].str.split('/',expand=True)
split_negara
```

```python
train_df.insert(1, 'Negara', split_negara[0])
train_df.insert(2, 'Tahun', split_negara[1])
```

```python
train_df.drop(columns=train_df.columns[0], inplace=True)
```

```python
train_df.head()
```

```python
split_test=test_df['Negara/Tahun'].str.split('/',expand=True)
split_test
```

```python
test_df.insert(1, 'Negara', split_test[0])
test_df.insert(2, 'Tahun', split_test[1])
```

```python
test_df.drop(columns=test_df.columns[0], inplace=True)
```

```python
test_df.head()
```

```python
train_df['Tahun']=pd.to_numeric(train_df['Tahun'], errors='coerce')
```

```python
test_df['Tahun']=pd.to_numeric(test_df['Tahun'], errors='coerce')
```
